```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Generate random values
np.random.seed(42)
n = 1000
pct_smokers = 0.2

# Vars
flag_smoker = np.random.rand(n) < pct_smokers
age = np.random.normal(40, 10, n)
height = np.random.normal(170, 10, n)
weight = np.random.normal(70, 10, n)

# Dataframe
dados = pd.DataFrame({
    'age': age,
    'height': height,
    'weight': weight,
    'flag_smoker': flag_smoker
})

# Data for flag_smoker
dados['flag_smoker'] = dados['flag_smoker'].map({True: 'Smoker', False: 'Non-smoker'})

# Display the first few rows of the dataframe
dados.head()
```

```python
# Style
sns.set(style = 'ticks')

# lmplot
sns.lmplot(x = 'height',
           y = 'weight',
           data = dados,
           hue = 'flag_smoker',
           palette = ['tab:blue', 'tab:red'],
           height = 8)

# Labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight for smokers and no smokers')

# Remove the borders
sns.despine()

# Show the plot
plt.show()
```
