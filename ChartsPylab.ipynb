```python
from pylab import *
```

```python
# Line charts with 2 imgs

# Data
x = linspace(0, 5, 10)
y = x ** 2

# Plot
fig = plt.figure()

# Create the axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Major figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Major figure')

# Minor figure
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Minor figure');
```

```python
# Data
x = linspace(0,5,10)
y = x ** 2

# Plot into two subplots
fig, axes = plt.subplots(nrows = 1, ncols = 4)

# For loop to create the plots
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Title')


# Adjust the layout
plt.tight_layout()
```
