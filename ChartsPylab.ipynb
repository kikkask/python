```python
from pylab import *
```

```python
# Line charts with 2 imgs

# Data
x = linspace(0, 5, 10)
y = x ** 2

# Plot
fig = plt.figure()

# Create the axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Major figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Major figure')

# Minor figure
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Minor figure');
```

```python
# Data
x = linspace(0,5,10)
y = x ** 2

# Plot into two subplots
fig, axes = plt.subplots(nrows = 1, ncols = 4)

# For loop to create the plots
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Title')


# Adjust the layout
plt.tight_layout()
```

```python
# Data
xx = np.linspace(-0.75, 1., 100)
n = np.array([0, 1, 2, 3, 4, 5])

# Subplots
fig, axes = plt.subplots(1, 4, figsize=(12,3))

# Plot 1
axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color = "black")
axes[0].set_title("scatter")

# Plot 2
axes[1].step(n, n**2, lw = 2, color = "blue")
axes[1].set_title("step")

# Plot 3
axes[2].bar(n, n**2, align = "center", width = 0.5, alpha = 0.5, color = "magenta")
axes[2].set_title("bar")

# Plot 4
axes[3].fill_between(x, x**2, x**3, color = "green", alpha = 0.5)
axes[3].set_title("fill_between");
```
